```python
import pandas as pd
import numpy as np
```

```python
df_saved_tracks = pd.DataFrame()

added_ts_list = []
artist_list = []
title_list = []
id_list = []
duration_list = []
explicit_list = []
album_list = []


more_songs = True
more_albums = True
offset_index = 0

while more_albums:
    albums = sp.current_user_saved_albums(limit=50, offset=offset_index)

    
    for album in albums["items"]:
        track_list = ''
        #print(album)
        album_name = album["album"]["name"]
        added_at = album["added_at"]
        
        tracks = album["album"]["tracks"]
        for track in tracks["items"]:
            #print(song)
            # tracks ids for retrive audio features
            track_list += track["id"] + ","

            # song attributes
            added_ts_list.append(added_at)
            title_list.append(track["name"])
            id_list.append(track["id"])
            duration_list.append(track["duration_ms"])
            explicit_list.append(track["explicit"])
            album_list.append(album_name) 


            # all the artists
            artists = track["artists"]
            artists_name = ""
            for artist in artists:
                artists_name += artist["name"] + ","
            artist_list.append(artists_name[:-1])

        # get track features
        track_features = sp.audio_features(track_list[:-1])
        if (track_features):
            print(".", album_name)
            track_features = [{} if v is None else v for v in track_features]
            df_temp = pd.DataFrame(track_features)    
        
            df_saved_tracks = df_saved_tracks.append(df_temp)
        track_list = ""
    
    if albums["next"] == None:
        # break the loop
        more_albums = False
    else:
        offset_index += albums["limit"]
    #more_albums = False    

df_saved_tracks["id"] = id_list
df_saved_tracks["song_title"] = title_list
df_saved_tracks["album"] = album_list
df_saved_tracks["duration_ms"] = duration_list
df_saved_tracks["explicit"] = explicit_list
df_saved_tracks["artists"] = artist_list
df_saved_tracks["added_at"] = added_ts_list
        
    
df_saved_tracks.shape
```
